Given a set of cities and distances between every pair of cities, the problem is to find the
shortest possible route that visits every city exactly once and returns to the starting point. Like
any problem, which can be optimized, there must be a cost function. In the context of TSP,
total distance traveled must be reduced as much as possible.
Consider the below matrix representing the distances (Cost) between the cities. Find the shortest
possible route that visits every city exactly once and returns to the starting point.

```python
import sys
import numpy as np
```

```python

def getLowesCostCity(matRow,current_path):
  row=matRow.shape[0]
  min=sys.maxsize
  cityIndex = 0
  temp=0
  for i in range (0,row):
    if(min>matRow[i] and  i not in current_path and matRow[i]!=0):
      temp=i
      min=matRow[i]
      cityIndex=i

    if(len(current_path)==row-1 and min>matRow[i] and matRow[i]!=0):
      min=matRow[i]
      cityIndex=i
      break
  return min,cityIndex



def tsp(mat):
  totalcost=0
  current_path=[0]
  row,col=mat.shape
  for i in range(0,(row)):
    tempCost,cityIndex=getLowesCostCity(mat[i],current_path)
    current_path.append(cityIndex)
    totalcost+=tempCost
  return totalcost



matrix = np.array([[0, 10, 15, 20],[10, 0, 35, 25], [15, 35, 0, 30], [20,  25, 30, 0]])
tsp(matrix)
row,col=matrix.shape
current_path=[]
tempCost,cityIndex=getLowesCostCity(matrix[0],current_path)
cost=tsp(matrix)
print("Min cost: ",cost)
```

Implement DFS on graph and tree.

```python
# graph

def DFS(graph, node, visited):
    if node not in visited:
        print(node)
        visited.append(node)
        for neighbor in graph[node]:
            DFS(graph, neighbor, visited)

graph = {'A': ['B', 'C'],
         'B': ['A', 'D', 'E'],
         'C': ['A', 'F'],
         'D': ['B'],
         'E': ['B', 'F'],
         'F': ['C', 'E']}

DFS(graph, 'A', [])

```

```python
# Tree
def DFS(node):
    print(node.value)
    for child in node.children:
        DFS(child)

class TreeNode:
    def __init__(self, value, children=[]):
        self.value = value
        self.children = children

root = TreeNode(1, [TreeNode(2, [TreeNode(4), TreeNode(5)]), TreeNode(3)])
DFS(root)
```

Write a program to solve the 8-puzzle problem (you can use any algo mentioned in this class).

```python
BOARD_SIZE = 8

def is_attacking(x1, y1, x2, y2):
    return x1 == x2 or y1 == y2 or abs(x1 - x2) == abs(y1 - y2)

def is_valid(board):
    for i in range(BOARD_SIZE):
        for j in range(i + 1, BOARD_SIZE):
            if board[i] != None and board[j] != None and is_attacking(i, board[i], j, board[j]):
                return False
    return True

def place_queens(board, row):
    if row == BOARD_SIZE:
        return True
    for col in range(BOARD_SIZE):
        board[row] = col
        if is_valid(board) and place_queens(board, row + 1):
            return True
    board[row] = None
    return False

def print_board(board):
    for row in range(BOARD_SIZE):
        for col in range(BOARD_SIZE):
            if board[row] == col:
                print("Q", end=" ")
            else:
                print(".", end=" ")
        print()

board = [None] * BOARD_SIZE
place_queens(board, 0)
print_board(board)
```

Implement Bi-directional search algorithm on the graph below.

```python
def bidirectional_search(graph, start, end):
    start_queue = [start]
    end_queue = [end]

    start_visited = set()
    end_visited = set()

    start_visited.add(start)
    end_visited.add(end)

    while start_queue and end_queue:
        if len(start_queue) <= len(end_queue):
            current = start_queue.pop(0)
            for neighbor in graph[current]:
                if neighbor in end_visited:
                    return "Path found"
                elif neighbor not in start_visited:
                    start_visited.add(neighbor)
                    start_queue.append(neighbor)
        else:
            current = end_queue.pop(0)
            for neighbor in graph[current]:
                if neighbor in start_visited:
                    return "Path found"
                elif neighbor not in end_visited:
                    end_visited.add(neighbor)
                    end_queue.append(neighbor)
    return "No path found"


graph = {
    'A': ['B', 'D'],
    'B': ['A', 'C', 'D'],
    'C': ['B', 'E'],
    'D': ['A', 'B', 'E'],
    'E': ['C', 'D']
}

start = 'A'
end = 'E'

result = bidirectional_search(graph, start, end)

print(result)  # Outputs "Path found"
```
