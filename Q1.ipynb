```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    'size': [1500, 2000, 1200, 1800, 2200, 1400, 1600, 2400, 1900, 1700],
    'bedrooms': [3, 4, 2, 3, 4, 2, 3, 4, 3, 2],
    'year_built': [1990, 1985, 2000, 1970, 1988, 1995, 2005, 1975, 1998, 1980],
    'price': [250000, 300000, 200000, 275000, 350000, 225000, 275000, 400000, 325000, 250000]
})

X = data.iloc[:, :-1].values  
y = data.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)




```

```python
score = model.score(X_test, y_test)
print(f'R^2 score: {score:.2f}')
```

```python
print(X_train)
print(y_train)
```

```python
# visualize training set result
import matplotlib.pyplot as plt

# Predict the prices on the test set
plt.scatter(X_train[:, 0], y_train, color='blue')
plt.plot(X_train[:, 0], model.predict(X_train), color='red')
plt.title('Training set')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (USD)')
plt.show()
```

```python
plt.scatter(X_test[:,0],y_test,color='blue')
plt.plot(X_test[:, 0], model.predict(X_test), color='red')
plt.title('Training set')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# Predict the price of a house with 1800 sq ft, 3 bedrooms, and built in 1995
house_features = pd.DataFrame({
    'size': [1800],
    'bedrooms': [3],
    'year_built': [1995]
})
predicted_price = model.predict(house_features.values)
print(f'Predicted price: ${predicted_price[0]:,.0f}')
```
