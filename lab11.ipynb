```python
import gym
import numpy as np

# Create the OpenAI gym environment
env = gym.make('FrozenLake-v1')

# Initialize the Q-table with zeros
Q = np.zeros([env.observation_space.n, env.action_space.n])

# Set hyperparameters
alpha = 0.8  # Learning rate
gamma = 0.95  # Discount factor
epsilon = 0.2  # Exploration rate
num_episodes = 5000  # Number of episodes to run

# Run the Q-learning algorithm
for episode in range(num_episodes):
    state = env.reset()
    done = False
    total_reward = 0

    while not done:
        # Choose an action using epsilon-greedy strategy
        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state, :])

        # Take the action and observe the next state and reward
        next_state, reward, done, _ = env.step(action)

        # Update the Q-table using the Q-learning formula
        Q[state, action] = (1 - alpha) * Q[state, action] + alpha * (reward + gamma * np.max(Q[next_state, :]))

        # Update the state and total reward
        state = next_state
        total_reward += reward

    # Print the total reward for this episode
    print(f"Episode {episode}: Total reward = {total_reward}")

# Print the final Q-table
print("Final Q-table:")
print(Q)
```

```python
import numpy as np

# Define the Q-value table
num_states = 10
num_actions = 4
Q = np.zeros((num_states, num_actions))

# Define the reinforcement learning parameters
alpha = 0.1
gamma = 0.9
epsilon = 0.1
num_episodes = 1000
max_steps_per_episode = 100

# Define the environment
def env_step(state, action):
    # Execute the action and observe the next state and reward
    if action == 0:
        next_state = max(0, state - 1)
    elif action == 1:
        next_state = min(num_states - 1, state + 1)
    elif action == 2:
        next_state = max(0, state - 2)
    else:
        next_state = min(num_states - 1, state + 2)
    reward = -1
    done = False
    if next_state == 0:
        reward = -100
        done = True
    elif next_state == num_states - 1:
        reward = 100
        done = True
    return next_state, reward, done

# Perform the reinforcement learning
for episode in range(num_episodes):
    # Reset the environment for the start of the episode
    state = np.random.randint(0, num_states)
    done = False

    for step in range(max_steps_per_episode):
        # Choose the next action using an epsilon-greedy policy
        if np.random.uniform() < epsilon:
            action = np.random.randint(0, num_actions)
        else:
            action = np.argmax(Q[int(state), :])

        # Execute the action and observe the next state and reward
        next_state, reward, done = env_step(state, action)

        # Update the Q-value table
        Q[int(state), action] += alpha * (reward + gamma * np.max(Q[int(next_state), :]) - Q[int(state), action])

        # Move to the next state
        state = next_state

        # End the episode if the terminal state is reached
        if done:
            print("Episode {} finished after {} timesteps with reward {}".format(episode + 1, step + 1, reward))
            break

    # Print the Q-value table at the end of the episode
    print("Q-value table after episode {}:".format(episode + 1))
    print(Q)
```
